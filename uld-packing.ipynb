```python
def createSequenceTable(A,B,C):
    n = len(A)
    dictA = {}
    for i in range(n):
        dictA[A[i]] = i+1
    
    dictB = {}
    for i in range(n):
        dictB[B[i]] = i+1

    dictC = {}
    for i in range(n):
        dictC[C[i]] = i+1
    
    table = [dictA, dictB, dictC]
    return table
```
